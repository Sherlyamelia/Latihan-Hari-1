```python
# Soal 1
def faktorial(n) :
    if n == 0 or n == 1 :
        return 1
    # Jika n adalah 0 atau 1, kembalikan 1.
    # Ini adalah kondisi dasar karena faktorial dari 0 dan 1 adalah 1. 
    
    else :
        return n * faktorial(n-1)
    # Jika n lebih besar dari 1, fungsi akan memanggil dirinya sendiri dengan nilai n-1, dan mengalikan hasilnya dengan n. 

# Memanggil fungsi faktorial dengan input 7 
print(faktorial(7))
```

```python
# Soal 2
# Fungsi untuk menghitung jumlah huruf vokal dalam kalimat
def hitung_vokal(kalimat) :
    vokal = 'aeiou'            # Daftar huruf vokal yang akan dihitung
    hasil = {}                 # Membuat kamus kosong untuk menyimpan hasil hitungan huruf vokal
    kalimat = kalimat.lower()  # Mengubah kalimat ke huruf kecil agar perhitungan tidak membedakan huruf besar dan kecil

    # Loop melalui setiap huruf vokal ('a','e','i','o','u')
    for v in vokal :
        jumlah = kalimat.count(v)   # Menghitung berapa kali huruf vokal muncul di kalimat
        
        # Jika huruf vokal setidaknya muncul sekali, simpah di kamus
        if jumlah > 0 :
            hasil[v] = jumlah  
    return hasil  # Mengembalikan kamus yang berisi jumlah kemunculan huruf vokal

kalimat = input("Masukkan sebuah kalimat")    # Input
hasil_vokal = hitung_vokal(kalimat)           # Memanggil fungsi hitung_vokal untuk menghitung jumlah vokal dalam kalimat
print("Jumlah huruf vokal dalam kalimat tersebut :", hasil_vokal) # Output dalam bentuk kamus
```

```python
# Soal 3
# Fungsi untuk mengambil angka genap dari sebuah daftar angka
def angka_genap(daftar) :

    # List comprehension digunakan untuk memilih angka genap
    # Menggunakan kondisi 'x % 2 == 0' untuk mengecek apakah angka genap
    return [x for x in daftar if x % 2 == 0]

daftar_angka = list(range(1,101))        # Membuat daftar angka dari 1 hingga 100 menggunakan fungsi range()
daftar_genap = angka_genap(daftar_angka) # Memanggil fungsi angka_genap untuk mendapatkan daftar angka genap dari daftar_angka
print(daftar_genap)
```

```python
# Soal 4
# Fungsi untuk memeriksa apakah sebuah string adalah palindrome
def cek_palindrome(s) :
    s = s.lower()            # Mengubah semua huruf menjadi huruf kecil agar tidak sensitif terhadap huruf besar/kecil
    s = s.replace(" "," ")   # Menghapus spasi dalma string untuk memastikan kata/kalimat yang dipisahkan spasi juga diperiksa dengan benar
    return s == s[::-1]      # Memeriksa apakah string sama jika dibaca dari depan maupun belakang

input_string = input("Masukkan sebuah kata atau kalimat :")  

# Memeriksa apakah string tersebut palindrome
if cek_palindrome(input_string) : 
    print(f'"{input_string}" adalah palindrome.')
else :
    print(f'"{input_string}" bukan palindrome.')
```

```python
# Soal 5
# Fungsi untuk mengurutkan daftar tuple berdasarkan elemen kedua
def urutkan_berdasarkan_elemen_kedua(daftar_tuple) :

    # Fungsi sorted() digunakan untuk mengurutkan, dengan key=lambda x: x[1]
    # Ini berarti mengurutkan berdasarkan elemen kedua dari setiap tuple (x[1])
    return sorted(daftar_tuple, key=lambda x: x[1])

daftar_tuple = [(1, 2), (3, 4), (5, 0)]
daftar_terurut = urutkan_berdasarkan_elemen_kedua(daftar_tuple)   # Memanggil fungsi untuk mengurutkan daftar
print(daftar_terurut)
```

```python
# Soal 1
import numpy as np 
rand_arr = np.random.randint(1, 1001, size=100)   # Membuat array NumPy 1D dengan 100 angka random antara 1 dan 1000
rata_rata = np.mean(rand_arr)   # Menghitung nilai rata-rata elemen array
median = np.median(rand_arr)    # Menghitung nilai median elemen array
std_dev = np.std(rand_arr)      # Menghitung standar deviasi elemen array
jumlah_lebih_besar_500 = np.sum(rand_arr > 500)  # Menghitung jumlah elemen yang lebih besar dari 500

# Menampilkan hasil jumlah perhitungan
print("Array Acak :", rand_arr)
print("Nilai Rata-rata :", rata_rata)
print("Nilai Median :", median)
print("Standar Deviasi :", std_dev)
print("Jumlah elemen lebih besar dari 500 :", jumlah_lebih_besar_500)
```

```python
# Soal 2
matrix = np.random.randint(0, 11, size=(4, 4))   # Mmebuat matriks 4x4
transpose_matrix = np.transpose(matrix)          # Mencari transpose
determinan = np.linalg.det(matrix)               # Mencari determinan

# Mencari invers jika determinan tidak nol
if determinan !=0 :
    invers_matrix = np.linalg.inv(matrix)
else :
    invers_matrix = "Invers tidak dapat dihitung"

# Menampilkan hasil
print("Matriks :\n", matrix)
print("Transpose :\n", transpose_matrix)
print("Determinan :", determinan)
print("Invers :\n", invers_matrix)
```

```python
# Soal 3
array = np.arange(1, 21)        # Membuat array dari 1 hingga 20
matrix = array.reshape(4, 5)   # Reshape menjadi 4x5

# Menghitung nilai maksimum setiap baris dan kolom
max_baris = np.max(matrix, axis = 1)
max_kolom = np.max(matrix, axis = 0)

# Menampilkan hasil
print("Matriks:\n", matrix)
print("Maksimum per baris :", max_baris)
print("Maksimum per kolom :", max_kolom)
```

```python
# Soal 4
# Membuat 2 matriks 3x3
matrix1 = np.random.randint(1, 10, size=(3, 3))
matrix2 = np.random.randint(1, 10, size=(3, 3))

penjumlahan = np.add(matrix1, matrix2)                   # Penjumlahan matriks
perkalian_elementwise = np.multiply(matrix1, matrix2)    # Perkalian element-wise
perkalian_matriks = np.dot(matrix1, matrix2)             # Perkalian matriks (dot product)

# Menampilkan hasil
print("Matriks 1:\n", matrix1)
print("Matriks 2:\n", matrix2)
print("Penjumlahan:\n", penjumlahan)
print("Perkalian element-wise:\n", perkalian_elementwise)
print("Perkalian matriks (dot product:\n", perkalian_matriks)
```
